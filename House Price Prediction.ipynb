Import the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the House Price Dataset

```python
house_price_dataset=sklearn.datasets.fetch_california_housing()
```

```python
#Loading the dataset to a pandas dataframe
house_price_dataframe=pd.DataFrame(house_price_dataset.data, columns=house_price_dataset.feature_names)
```

```python
#Printing the first 5 rows of dataframe
house_price_dataframe.head()
```

```python
#Adding the target label (price) to the dataset
house_price_dataframe['Price']=house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
#checking the number of rows and columns in the dataframe
house_price_dataframe.shape
```

```python
#Check for missing values
house_price_dataframe.isnull().sum()
```

```python
#Statistical Measures of the Dataset
house_price_dataframe.describe()
```

Understanding the correlation between various features of the dataset

```python
correlation=house_price_dataframe.corr()
```

```python
#Constructing a heatmap to understand the correlation
plt.figure(figsize=(9,9))
sns.heatmap(correlation,cbar=True,square=True,fmt=".1f",annot=True,annot_kws={'size':8},cmap='Blues')
```

Splitting the data and target(label)

```python
X=house_price_dataframe.drop(['Price'],axis=1)
Y=house_price_dataframe['Price']
```

```python
print(X)
print(Y)
```

Splitting the data into training data and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

MODEL TRAINING - XGBOOST Regression

```python
#Loading the model
model=XGBRegressor()
```

```python
#Training the model with training data
model.fit(X_train,Y_train)
```

Evaluation of the model

Prediction on training data

```python
#accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R Squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared Error : ",score_1)
print("Mean Absolute Error : ",score_2)
```

Visualizing the actual prices and predicted prices

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price VS Predicted Price")
plt.show()
```

Prediction on test data

```python
#accuracy for prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
# R Squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared Error : ",score_1)
print("Mean Absolute Error : ",score_2)
```
